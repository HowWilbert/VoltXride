```python
# ===============================
# 1️⃣ Import Libraries
# ===============================

import pandas as pd
import numpy as np
# !pip install statsmodels
import statsmodels.api as sm
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score

# ===============================
# 2️⃣ Load Dataset
# ===============================

file_path = "DECODEX_DAY3_ANALYSIS.xlsx"

zone_hour = pd.read_excel(file_path, sheet_name="Zone_Hour_Demand")
charging = pd.read_excel(file_path, sheet_name="Charging_Stations")

# ===============================
# 3️⃣ Feature Engineering
# ===============================

# Cancellation Rate
zone_hour["Cancellation_Rate"] = (
    zone_hour["Avg_Cancellations"] / zone_hour["Avg_Ride_Requests"]
)

# Convert Peak Flag to binary
zone_hour["Peak_Flag_Binary"] = zone_hour["Peak Flag"].apply(
    lambda x: 1 if x == "Peak" else 0
)

# Aggregate charger data at City-Zone level
chargers_zone = charging.groupby(["City", "Zone"]).agg({
    "Chargers_Available": "sum",
    "Avg_Wait_Time_Min": "mean"
}).reset_index()

# Merge datasets
data = pd.merge(zone_hour, chargers_zone,
                on=["City", "Zone"], how="left")

# Fill missing
data["Chargers_Available"] = data["Chargers_Available"].fillna(0)
data["Avg_Wait_Time_Min"] = data["Avg_Wait_Time_Min"].fillna(0)

# Chargers per 1000 rides
data["Chargers_per_1000"] = (
    data["Chargers_Available"] / data["Avg_Ride_Requests"]
) * 1000

# Interaction term (Peak × Wait Time)
data["Peak_Wait_Interaction"] = (
    data["Peak_Flag_Binary"] * data["Avg_Wait_Time_Min"]
)

# ===============================
# 4️⃣ Multivariate Linear Regression
# ===============================

X_linear = data[[
    "Peak_Flag_Binary",
    "Avg_Ride_Requests",
    "Hour",
    "Chargers_per_1000",
    "Avg_Wait_Time_Min",
    "Peak_Wait_Interaction"
]]

X_linear = sm.add_constant(X_linear)
y = data["Cancellation_Rate"]

linear_model = sm.OLS(y, X_linear).fit()

print("Linear Regression Summary")
print(linear_model.summary())

# Extract R²
linear_r2 = linear_model.rsquared
print("\nLinear Model R²:", round(linear_r2, 3))


# ===============================
# 5️⃣ Random Forest (Nonlinear Model)
# ===============================

features = [
    "Peak_Flag_Binary",
    "Avg_Ride_Requests",
    "Hour",
    "Chargers_per_1000",
    "Avg_Wait_Time_Min"
]

rf = RandomForestRegressor(
    n_estimators=300,
    random_state=42
)

rf.fit(data[features], y)

rf_predictions = rf.predict(data[features])
rf_r2 = r2_score(y, rf_predictions)

print("\nRandom Forest R²:", round(rf_r2, 3))


# ===============================
# 6️⃣ Scenario Simulation
# Reduce Peak Wait Time by 20%
# ===============================

scenario = data.copy()

# Reduce wait time only during peak
scenario.loc[
    scenario["Peak_Flag_Binary"] == 1,
    "Avg_Wait_Time_Min"
] *= 0.8

# Recalculate interaction
scenario["Peak_Wait_Interaction"] = (
    scenario["Peak_Flag_Binary"] *
    scenario["Avg_Wait_Time_Min"]
)

scenario_X = sm.add_constant(scenario[[
    "Peak_Flag_Binary",
    "Avg_Ride_Requests",
    "Hour",
    "Chargers_per_1000",
    "Avg_Wait_Time_Min",
    "Peak_Wait_Interaction"
]])

pred_current = linear_model.predict(X_linear)
pred_scenario = linear_model.predict(scenario_X)

avg_current = np.mean(pred_current)
avg_scenario = np.mean(pred_scenario)

reduction = avg_current - avg_scenario

print("\nPredicted Avg Cancellation (Current):",
      round(avg_current, 4))
print("Predicted Avg Cancellation (After 20% Peak Wait Reduction):",
      round(avg_scenario, 4))
print("Estimated Reduction:",
      round(reduction, 4))


# ===============================
# 7️⃣ Charger Density Impact Simulation
# Increase chargers per 1000 by 1 unit
# ===============================

scenario_charger = data.copy()
scenario_charger["Chargers_per_1000"] += 1

scenario_charger_X = sm.add_constant(scenario_charger[[
    "Peak_Flag_Binary",
    "Avg_Ride_Requests",
    "Hour",
    "Chargers_per_1000",
    "Avg_Wait_Time_Min",
    "Peak_Wait_Interaction"
]])

pred_charger = linear_model.predict(scenario_charger_X)

charger_reduction = np.mean(pred_current) - np.mean(pred_charger)

print("\nEstimated Cancellation Reduction from +1 Charger per 1000:",
      round(charger_reduction, 4))
```

```python
import pandas as pd
import numpy as np

from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("fleet_redploy.csv")

# Cancellation Rate (Target Variable)
df["Cancellation_Rate"] = (
    df["Total_Requests"] - df["Completed_Rides"]
) / df["Total_Requests"]

# Demand share inside city (important feature)
df["Zone_Demand_Share"] = (
    df["Total_Requests"] /
    df.groupby("City")["Total_Requests"].transform("sum")
)
```

```python
# ML FEATURES 
features = [
    "Fleet_Surplus_Index",
    "Total_Requests",
    "City_Completion_Rate",
    "Expected_Completions",
    "Zone_Demand_Share"
]

X = df[features]
y = df["Cancellation_Rate"]
```

```python
#Train MODEL 
model = RandomForestRegressor(
    n_estimators=200,
    random_state=42
)

model.fit(X, y)
```

```python
#CORE ENGINE 
def predict_cancellation(temp_df, model, features):
    X_temp = temp_df[features]
    return model.predict(X_temp)
```

```python
#MAIN LOGIC 
def reallocation_engine(df, city_name, model, features,
                        transfer_unit=1,
                        safety_buffer=1):

    city_df = df[df["City"] == city_name].copy()

    surplus_zones = city_df[
        city_df["Fleet_Surplus_Index"] > safety_buffer
    ]

    deficit_zones = city_df[
        city_df["Fleet_Surplus_Index"] < 0
    ]

    recommendations = []

    base_pred = predict_cancellation(city_df, model, features)
    base_total_cancel = np.sum(
        base_pred * city_df["Total_Requests"]
    )

    for s_idx, s_row in surplus_zones.iterrows():
        for d_idx, d_row in deficit_zones.iterrows():

            temp_df = city_df.copy()

            # simulate movement
            temp_df.loc[s_idx, "Fleet_Surplus_Index"] -= transfer_unit
            temp_df.loc[d_idx, "Fleet_Surplus_Index"] += transfer_unit

            new_pred = predict_cancellation(temp_df, model, features)

            new_total_cancel = np.sum(
                new_pred * temp_df["Total_Requests"]
            )

            improvement = base_total_cancel - new_total_cancel

            if improvement > 0:
                recommendations.append({
                    "City": city_name,
                    "From_Zone": s_row["Pickup_Zone"],
                    "To_Zone": d_row["Pickup_Zone"],
                    "Fleet_Shifted": transfer_unit,
                    "Cancellation_Reduction": improvement
                })

    rec_df = pd.DataFrame(recommendations)

    if len(rec_df) > 0:
        rec_df = rec_df.sort_values(
            "Cancellation_Reduction",
            ascending=False
        )

    return rec_df
```

```python
#MUMBAI EXAMPLE 
mumbai_reallocation = reallocation_engine(
    df,
    city_name="Bengaluru",
    model=model,
    features=features,
    transfer_unit=50

)

print(mumbai_reallocation.head(10))
```

```python
# ============================================================
# VOLTRIDE FLEET REALLOCATION IMPACT MODEL
# Objective:
# Measure reduction in COMPANY cancellation rate
# when fleet is reallocated ONLY within Mumbai.
# ============================================================

import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor

# ============================================================
# STEP 1 — LOAD DATA
# ============================================================

# Replace with your file name
df = pd.read_csv("fleet_redploy.csv")

# ============================================================
# STEP 2 — CREATE TARGET VARIABLE
# ============================================================

# Cancellation Rate (Target for ML model)
df["Cancellation_Rate"] = (
    df["Total_Requests"] - df["Completed_Rides"]
) / df["Total_Requests"]

# Demand share inside city (important feature)
df["Zone_Demand_Share"] = (
    df["Total_Requests"] /
    df.groupby("City")["Total_Requests"].transform("sum")
)

# ============================================================
# STEP 3 — ML FEATURES
# ============================================================

features = [
    "Fleet_Surplus_Index",
    "Total_Requests",
    "City_Completion_Rate",
    "Expected_Completions",
    "Zone_Demand_Share"
]

X = df[features]
y = df["Cancellation_Rate"]

# ============================================================
# STEP 4 — TRAIN MODEL
# ============================================================

model = RandomForestRegressor(
    n_estimators=200,
    random_state=42
)

model.fit(X, y)

# ============================================================
# STEP 5 — COMPANY CANCELLATION RATE FUNCTION
# ============================================================

def company_cancellation_rate(dataframe, predicted_cancel_rate):

    cancelled_rides = (
        predicted_cancel_rate *
        dataframe["Total_Requests"]
    )

    total_cancelled = cancelled_rides.sum()
    total_requests = dataframe["Total_Requests"].sum()

    return total_cancelled / total_requests


# ============================================================
# STEP 6 — BASE COMPANY CANCELLATION RATE
# ============================================================

base_pred = model.predict(df[features])

base_company_rate = company_cancellation_rate(df, base_pred)

print("\nBase Company Cancellation Rate:",
      round(base_company_rate * 100, 3), "%")


# ============================================================
# STEP 7 — MUMBAI FLEET REALLOCATION
# Example:
# Move fleet from Mumbai Zone 10 → Mumbai Zone 8
# ============================================================

df_after = df.copy()

TRANSFER_UNIT = 1   # number of vehicles shifted

# FROM ZONE (SURPLUS)
df_after.loc[
    (df_after["City"] == "Mumbai") &
    (df_after["Pickup_Zone"] == 10),
    "Fleet_Surplus_Index"
] -= TRANSFER_UNIT

# TO ZONE (DEFICIT)
df_after.loc[
    (df_after["City"] == "Mumbai") &
    (df_after["Pickup_Zone"] == 8),
    "Fleet_Surplus_Index"
] += TRANSFER_UNIT


# ============================================================
# STEP 8 — NEW COMPANY CANCELLATION RATE
# ============================================================

new_pred = model.predict(df_after[features])

new_company_rate = company_cancellation_rate(
    df_after,
    new_pred
)

print("New Company Cancellation Rate:",
      round(new_company_rate * 100, 3), "%")


# ============================================================
# STEP 9 — FINAL IMPACT
# ============================================================

reduction = base_company_rate - new_company_rate

print("\nReduction in Company Cancellation Rate:",
      round(reduction * 100, 3), "%")


# ============================================================
# OPTIONAL — IMPROVEMENT IN NUMBER OF RIDES SAVED
# ============================================================

base_cancelled = (base_pred * df["Total_Requests"]).sum()
new_cancelled = (new_pred * df_after["Total_Requests"]).sum()

rides_saved = base_cancelled - new_cancelled

print("Estimated Reduction in Cancelled Rides:",
      round(rides_saved, 2))

```
